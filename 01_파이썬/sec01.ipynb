```python
# 참조건
print(10 > 5)
print(10<5)
print(10 >= 10)
print(10 <=10)
print(10==10)
print(10!=10)
print(10=10)
```

```python
# 조건문
score = 90
if score >= 90:
    print("A")
else:
    print("B")
```

```python
점수 = 87
if 점수 >= 90:
    print("a")
elif 점수 >= 80:
    print("b")
elif 점수 >= 70:
    print("c")
else:
    print("f")
```

```python
# 조건문의 구성
score = 2
if score >= 10:
    print("score가 10이상 입니다.")
elif score < 10 and score > 5:
    print("score가 5이상 10미만 입니다.")
elif score < 5  and score >= 3:
    print("score가 3이상 5미만 입니다.")
else:
    print("score가 3미만 입니다.")
```
